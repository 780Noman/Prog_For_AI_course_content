```python
import ChatBot
from chatterbot import ChatBot
from chatterbot.trainers import ChatterBotCorpusTrainer
from chatterbot.trainers import ListTrainer
from textblob import TextBlob  # Import TextBlob for sentiment analysis

# Create a chatbot instance
mentor_bot = ChatBot('AiMentor')

# Create a new trainer for the chatbot
trainer = ChatterBotCorpusTrainer(mentor_bot)

# Train the chatbot on English language data
trainer.train('chatterbot.corpus.english')

# Custom dataset for additional training
custom_dataset = [
    "Tell me about machine learning.",
    "Machine learning is a field of artificial intelligence that focuses on building systems that can learn from data.",
    "How can I start learning programming?",
    "To start learning programming, choose a language like Python, and explore online resources like Codecademy or freeCodeCamp.",
    "Give me some motivational quotes.",
    "Sure, here's one: 'The only way to do great work is to love what you do.' - Steve Jobs",
    "What are some career development tips?",
    "Focus on continuous learning, build a strong network, and set clear career goals to guide your path.",
    "Can you recommend a book on artificial intelligence?",
    "Certainly! 'Artificial Intelligence: A Modern Approach' by Stuart Russell and Peter Norvig is a great resource.",
    "What online courses do you recommend for learning programming?",
    "For programming, consider courses on platforms like Coursera, edX, and Udacity. Popular ones include 'Python for Everybody' and 'JavaScript: The Complete Guide'.",
    "Which programming language should I learn first?",
    "If you're a beginner, Python is a great choice due to its readability and versatility. If you're interested in web development, consider starting with HTML, CSS, and JavaScript."
]

# Train the chatbot on the custom dataset
custom_trainer = ListTrainer(mentor_bot)
custom_trainer.train(custom_dataset)

# Simple interaction loop with AI mentor features
while True:
    user_input = input("You: ")

    # Exit the loop if the user enters 'exit'
    if user_input.lower() == 'exit':
        print("AiMentor: Goodbye! Keep learning and stay curious.")
        break

    # Sentiment analysis to determine user's mood
    analysis = TextBlob(user_input)
    sentiment = "positive" if analysis.sentiment.polarity > 0 else "negative" if analysis.sentiment.polarity < 0 else "neutral"

    # Handling specific queries and providing guidance
    if "learn programming" in user_input.lower():
        response = "To start learning programming, choose a language like Python, and explore online resources like Codecademy or freeCodeCamp."

    elif "machine learning" in user_input.lower():
        response = "Machine learning is a field of artificial intelligence that focuses on building systems that can learn from data."

    elif "motivational" in user_input.lower():
        response = "Sure, here's a motivational quote: 'The only way to do great work is to love what you do.' - Steve Jobs"

    elif "career development" in user_input.lower():
        response = "Focus on continuous learning, build a strong network, and set clear career goals to guide your path."

    elif "book on artificial intelligence" in user_input.lower():
        response = "Certainly! 'Artificial Intelligence: A Modern Approach' by Stuart Russell and Peter Norvig is a great resource."

    elif "online courses for learning programming" in user_input.lower():
        response = "For programming, consider courses on platforms like Coursera, edX, and Udacity. Popular ones include 'Python for Everybody' and 'JavaScript: The Complete Guide'."

    elif "programming language to learn first" in user_input.lower():
        response = "If you're a beginner, Python is a great choice due to its readability and versatility. If you're interested in web development, consider starting with HTML, CSS, and JavaScript."

    elif sentiment == "positive":
        response = "That's fantastic! Keep up the good work and stay motivated."

    elif sentiment == "negative":
        response = "I'm here to help. Remember, challenges are opportunities in disguise. Keep pushing forward!"

    else:
        # Get the chatbot's response for general queries
        response = mentor_bot.get_response(user_input)

    print("AiMentor:", response)
```
